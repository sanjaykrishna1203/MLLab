```python
import pandas as pd

data = pd.read_csv("salary.csv")
x = list(data['YearsExperience'])
y = list(data['Salary'])
print(data)
```

```python
def linear_regression(x,y):
    sumX = sum(x)
    sumY = sum(y)
    meanX = sumX/len(x)
    meanY = sumY/len(y)
    xMinusXMean = [i-meanX for i in x]
    yMinusYMean = [i-meanY for i in y]
    zipLi = zip(xMinusXMean, yMinusYMean)
    val = [x*y for x,y in zipLi]
    b1 = sum(val)/sum([x**2 for x in xMinusXMean])
    b0 = meanY - b1 * meanX
    return b0, b1
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.5, shuffle='true')
b = linear_regression(x_train, y_train)
y_pred = [b[0]+b[1]*val for val in x_test]

import  matplotlib.pyplot as plt
plt.plot(x_test, y_pred)
plt.scatter(x_test, y_test, c="k")

```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
import numpy as np

print("MSE ", np.sqrt(mean_squared_error(y_test, y_pred)))
print("r2 score ", r2_score(y_test, y_pred))
```
